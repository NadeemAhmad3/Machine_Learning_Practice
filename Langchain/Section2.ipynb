```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage
prompt=ChatPromptTemplate.from_messages([
        ("system","Act as an {role},and answer the {question}"),
        ("human","in {language}")
    
])
chain = prompt | llm
response=chain.invoke ({"role":"GenAI Instructor","question":"benefit of genai in 2026","language":"URDU"})
print(response.content)
## working with history now 
hisprompt=ChatPromptTemplate.from_messages([
    ("system","You are a {input} instructor"),
    MessagesPlaceholder(variable_name="history"),
    ("human","Benefit of studying {subject}")
     
])
chain1= hisprompt | llm
response1=chain1.invoke({"input":"data science","subject":"data science","history":[]})
print(response1.content)
## use that history now
print("using history now")
conversation_history = [
    HumanMessage(content="Benefit of studying data science"),   
    AIMessage(content=response1.content)                        
]
response2 = chain1.invoke({
    "input": "data science",              
    "subject": "it",                    
    "history": conversation_history       
})

print("=== WITH HISTORY ===")
print(response2.content)
```
